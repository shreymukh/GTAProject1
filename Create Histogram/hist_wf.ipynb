```python
import pandas as pd
import matplotlib.pyplot as plt

# Read csv
wf_df = pd.read_csv('../top 100 cities_whole foods.csv')
wf_df = wf_df.sort_values(by=['# of Whole Foods (15km radius)'],ascending=False)
```

```python
states = []
wf = []
count = 0
for item in wf_df['State']:
    if item not in states:
        states.append(item)
    index = states.index(item)
    wf[index] += wf_df['# of Whole Foods (15km radius)'][count]
    count +=1
```

```python
states = ['Texas', 'Maine', 'Georgia']
states.index('Maine')
```

```python
myhist = wf_df.plot(kind = 'bar', x = 'State', y = '# of Whole Foods (15km radius)', figsize = (20,10)).set_ylabel('# of Whole Foods\n')
fig = myhist.get_figure()
plt.savefig('../WF_Histogram.png', bbox_inches = 'tight')      ('Median Sales Price ($)')
```
