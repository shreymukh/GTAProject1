```python
import csv
import pandas as pd       
from re import sub
from decimal import Decimal
```

```python
with open('../Median_Sale_Rental.csv', newline = '') as csvfile1:
    csvreader1 = csv.reader(csvfile1, delimiter=',')
    csv_header1 = next(csvreader1)

    city = []
    state = []
    med_Sales = []
    med_Rent = []
    numWF = []
    rank = []
    
    for row in csvreader1:
        my_city = row[2]
        state.append(row[3])
        
        if row[4] == "":
            break
        med_Sales.append(row[4])

        med_Rent.append(row[5])
        
        with open('../top 100 cities_whole foods.csv', newline = '') as csvfile2:
            csvreader2 = csv.reader(csvfile2, delimiter=',')
            csv_header2 = next(csvreader2)
                         
            my_num = "N/A"
            my_rank = "N/A"
            for my_row in csvreader2:
                if row[2].find(my_row[2]) == 0:
                    my_city = my_row[2]
                    my_num = my_row[6]
                    my_rank = int(my_row[1])
                    
            numWF.append(my_num)
            rank.append(my_rank)
            city.append(my_city)
                        
csvfile1.close()
csvfile2.close()

with open('../cities_housing_wf.csv', 'w', newline='') as csvfile3:
    csvwriter3 = csv.writer(csvfile3, delimiter=',')
    csvwriter3.writerow(['Rank', 'City', 'State', 'Median Sales Price', 'Median Rental', '# of Whole Foods'])
                        
    # write lists to rows
    for i in range(len(rank)):
        csvwriter3.writerow([rank[i], city[i], state[i], med_Sales[i], med_Rent[i], numWF[i]])                            
                         
```

```python
cities_df = pd.read_csv('../cities_housing_wf.csv')
cities_df = cities_df.loc[cities_df['Rank'] >= 0,:]
cities_df = cities_df.sort_values(by=['Rank'])
cities_df = cities_df.reset_index(drop=True)
cities_df
```
