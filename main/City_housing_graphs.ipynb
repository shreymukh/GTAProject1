```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import csv
```

```python
# Load in csv
city_df = pd.read_csv("C:/Users/17703/Shreya/GTAProject1/main/top100_median_with_wf.csv")
city_df.head()
```

```python
#Median Home Sales Price VS Cities

city_df.plot(kind = "bar", x = "City", y= "Median Sale", figsize =(20,10)).set_ylabel("Median Sales Price($)")
plt.title("Median Home Sales Price in US Cities")
plt.xlabel("Cities")
plt.ylabel("Median Sales Price")
plt.tight_layout()
plt.savefig("Median Home Sales Price.png")
plt.show()
```

```python
 #Median Home Rents VS Cities

city_df.plot(kind = "bar", x = "City", y= "Median Rent", figsize =(20,10)).set_ylabel("Median Rental($)")
plt.title("Median Rents VS US Cities")
plt.xlabel("Cities")
plt.ylabel("Median Rent")
plt.tight_layout()
plt.savefig("Median Home Rent.png")
plt.show()
```

```python
# Whole Foods count within 15 km radius per city

city_df.plot(kind = "bar", x = "City", y= "# of Whole Foods (15km radius)", figsize =(20,10)).set_ylabel("Median Sales Price($)")
plt.title("Count of Whole Foods within 15 km radius VS US Cities")
plt.xlabel("Cities")
plt.ylabel("Whole Foods within 15 km radius)")
plt.tight_layout()
plt.savefig("Citywise_Wholefoods.png")
plt.show()
```

```python
# Count of Whole foods across US Cities

city_df.plot(kind = "bar", x = "City", y= "# of Whole Foods (15km radius)", color = "g", figsize =(20,10)).set_ylabel("Whole Foods)")
plt.title("No. of Whole Foods Across US Cities")
plt.xlabel("Cities")
plt.ylabel("Whole Foods(15km radius)")
plt.tight_layout()
plt.savefig("Whole Food Cities.png")
plt.show()
```

```python
# Whole Foods count and Median Sales price per city

city_df.plot(kind = "bar", x = "Median Sale", y= "# of Whole Foods (15km radius)", color = "c", figsize =(20,10)).set_ylabel("Whole Foods)")
plt.title("No. of Whole Foods Across US Cities")
plt.xlabel("Median Sale")
plt.ylabel("Whole Foods(15km radius)")
plt.tight_layout()
plt.savefig("WholeFoods_MedianSale.png")
plt.show()
```

```python
# Whole Foods count and Median Sales price per city

city_df.plot(kind = "bar", x = "Median Rent", y= "# of Whole Foods (15km radius)", color = "c", figsize =(20,10)).set_ylabel("Whole Foods)")
plt.title("No. of Whole Foods Across US Cities")
plt.xlabel("Median Rent")
plt.ylabel("Whole Foods(15km radius)")
plt.tight_layout()
plt.savefig("WholeFoods_MedianRent.png")
plt.show()
```
